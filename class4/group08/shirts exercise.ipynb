```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({'gender': [1, 1, 1, 0, 0],
                     'color': ['bright blue', 'navy blue', 'navy blue', 'light blue', 'yellow'],
                     'sleeve': [-1, 1, 0, 0, 1],
                     'pattern': [1, 0, 1, 0, 1],
                     'casual': [0, 0, 1, 1, 1]})
```

```python
data
```

```python
dummy = pd.get_dummies(data['color'])
```

```python
data = data.join(dummy)
data
```

```python
data = data.drop(['color'], axis=1)
data
```

```python
def cosine_similarity(a,b):
    dot_product = np.dot(a,b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)
```

```python
length = data.shape[0]
matrix = [[1 for i in range(length)] for j in range(length)]
for i in range(length):
    for j in range(i+1, length):
        matrix[i][j] = cosine_similarity(data.loc[i], data.loc[j])
        matrix[j][i] = cosine_similarity(data.loc[i], data.loc[j])
```
